```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split,KFold,cross_val_score
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.metrics import accuracy_score
import numpy as np
```

```python
data = pd.read_csv("bank-additional-full.csv")
data.head(30)
```

```python
data = pd.concat([data, pd.get_dummies(data["marital"],prefix="marital")], axis=1)
```

```python
data= pd.concat([data, pd.get_dummies(data["housing"],prefix="housing")], axis=1)
```

```python
data= pd.concat([data, pd.get_dummies(data["contact"],prefix="contact")], axis=1)
```

```python
data= pd.concat([data, pd.get_dummies(data["poutcome"],prefix="poutcome")], axis=1)
```

```python
data.drop(["marital","housing","contact","poutcome"],axis = 1, inplace= True)
```

```python
data.head()
```

```python
data.drop(["job","education","default","month","day_of_week","pdays"],axis = 1, inplace= True)
```

```python
data.head()
```

```python
data.drop(["marital_unknown","housing_unknown","housing_no","contact_cellular","poutcome_failure","poutcome_nonexistent"], axis = 1 , inplace = True)
```

```python
data.head()
```

```python
data = pd.concat([data,pd.get_dummies(data["loan"], prefix="loan")], axis = 1)
```

```python
data = pd.concat([data,pd.get_dummies(data["y"], prefix="y")], axis = 1)
```

```python
data.head()
```

```python
data.drop(["loan_no","loan_unknown","y_no"],axis = 1 , inplace = True)
```

```python
data.head()
```

```python
data.drop(["loan","y"], axis=1, inplace = True)
```

```python
data.head()
```

```python
y=data["y_yes"]
```

```python
data.drop(["y_yes"], axis=1, inplace= True)
```

```python
X = data.iloc[:,:16]
```

```python
X.head()
```

Split The Data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1234)
model  = LogisticRegression()
```

```python
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy_score(y_test, y_pred)
print('Accuracy score for random_state = : ' + str(accuracy_score(y_test, y_pred)))
```

```python
# K-fold split

folds = KFold(n_splits = 5, shuffle = True, random_state = 35)
scores = []

for n_fold, (train_index, valid_index) in enumerate(folds.split(X,y)):
    print('\n Fold '+ str(n_fold+1 ) + 
          ' \n\n train ids :' +  str(train_index) +
          ' \n\n test ids :' +  str(valid_index))
    
    X_train, X_valid = X.iloc[list(train_index)], X.iloc[list(valid_index)]
    y_train, y_valid = y.iloc[list(train_index)], y.iloc[list(valid_index)]
    
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    acc_score = accuracy_score(y_test, y_pred)
    scores.append(acc_score)
    print('\n Accuracy score for Fold ' +str(n_fold+1) + ' --> ' + str(acc_score)+'\n')

    
print(scores)
print('Avg. accuracy score :' + str(np.mean(scores)))
```
